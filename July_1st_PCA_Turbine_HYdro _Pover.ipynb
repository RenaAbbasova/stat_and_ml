```python
data = pd.read_csv('turbine_hydro_power.csv')
```

```python
df=data.copy()
```

```python
df.head()
```

```python
df=df.drop(["Unnamed:0"],axis=1)
```

```python
X_scaled.head()
```

```python
Y_predict=RF.predict(X_test)
```

```python
# crosscheck performance
```

```python
# understand the PCA concept
```

```python
# understand the Eigen vestor,eigen value,principal component
```

```python
# step1---calculculating the covariance matrix
```

```python
cov_matrix=X_scaled.cov()
```

```python
cov_matrix
```

```python
cov_matrix.shape
```

```python
# calculate the eigen value,eigen vector
```

```python
# numpy.linalg.eig()--will return eigen value and eigen vector
```

```python
eigen_value,eigen_vector=np.linalg.eig(cov_matrix)
```

```python
eigen_value.shape
```

```python
#show many eigen value--same no. of columns
len(eigen_value)# reprecented by columns# eigen value summarizacion of the variance
```

```python
len(eigen_vector)#reprecented by columns
```

```python
# calculation of the principal component data is the the dot product of original data, eigen vector
```

```python
principal_components_data=np.dot(X_scaled,eigen_vector)
principal_components_data
```

```python
# applying PCA is easy..
```

```python
pca10=PCA()
```

```python
X_scaled.shape
```

```python
pca10=PCA()
```

```python
X_pca=pca10.fit(X_scaled)
```

```python
X_pca
```

```python
X_pca.explained_variance_ratio_
```

```python
np.sum(X_pca.explained_variance_ratio_)
```

```python
len(X_pca.explained_variance_ratio_)
```

```python
# visualize of information with respect to PCA number selection
```

```python
plt.figure(figsize=(10,6))
plt.plot(np.cumsum(X_pca.explained_variance_ratio_))
plt.xlabel("no. of principal components")
plt.ylabel("amount of information we received")
plt.show()
```

```python
# we are able to extract upto 95% information with 150 principal components
```
