```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Energy_and_Water_Data.csv')
```

```python
pd.set_option('display.max_rows',100)
```

```python
df=data.copy()
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop('Order',axis=1,inplace=True)
```

```python
col=['Latitude','Longitude','Community Board','Council District','Census Tract','DOF Gross Floor Area']                                                  
                                                
```

```python
df[col]=df[col].fillna(df[col].mean())
```

```python
df['Water Required?']=df['Water Required?'].fillna(df['Water Required?'].mode()[0])
```

```python
df['DOF Benchmarking Submission Status']=df['DOF Benchmarking Submission Status'].fillna(df['DOF Benchmarking Submission Status'].mode()[0])
```

```python
df['NTA']=df['NTA'].fillna(df['NTA'].mode()[0])
```

```python
df['BBL - 10 digits']=df['BBL - 10 digits'].fillna(df['BBL - 10 digits'].mode()[0])
```

```python
df['Borough']=df['Borough'].fillna(df['Borough'].mode()[0])
```

```python
df['Street Number']=df['Street Number'].fillna(df['Street Number'].mode()[0])
```

```python
df['Street Name']=df['Street Name'].fillna(df['Street Name'].mode()[0])
```
