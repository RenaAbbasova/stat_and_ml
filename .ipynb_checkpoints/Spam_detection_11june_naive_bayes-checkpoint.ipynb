```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('spam.csv',encoding='latin-1')
```

```python
data.head()
```

```python
df=data.copy()
```

```python
df = df[['v2', 'v1']].rename(columns={'v2': 'email', 'v1': 'label'})
```

```python
df.head()
```

```python
# split into dependet and independet variable
```

```python
X=df['email']
Y=df['label']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10,stratify=Y)
```

```python
X_train.shape
```

```python
X_train.head()
```

```python
# basic concept of NLP
# bag of word creation--some libraries help to create
# countvectorizer, Ifidvectorizer
```

```python

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
vector=TfidfVectorizer()
```

```python

vectorized_data = vector.fit_transform(X_train)
```

```python
new_X_train=vector.transform(X_train)
```

```python
X_train_vector=vector.transform(X_test)
```

```python
X_train_vector.shape
```

```python
count_array=X_train_vector.toarray()
count_array
```

```python
# lets see the bag of words in dataframe format
```

```python
df100 = pd.DataFrame(data=count_array, columns=vector.get_feature_names_out())
```
