```python
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
# creation of dependent and independent variable
X,Y=load_breast_cancer(return_X_y=True)
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
# create training testing dataset

```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=10)
```

```python
X.shape
```

```python
X_train.shape
```

```python
# model bulding process
from sklearn.tree import DecisionTreeClassifier,DecisionTreeRegressor
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
dt=DecisionTreeClassifier(criterion='gini',class_weight='balanced')
```

```python
dt.fit(X_train,Y_train)
```

```python
y_train_predict=dt.predict(X_train)
```

```python
accuracy_score(Y_train,y_train_predict)
```

```python
y_test_predict=dt.predict(X_test)
```

```python
accuracy_score(Y_test,y_test_predict)
```

```python
from sklearn import tree
```

```python
plt.figure(figsize=(20,12))
tree.plot_tree(dt)
plt.show()
```

```python
# post pruning
```

```python
dt1=DecisionTreeClassifier(criterion='entropy',class_weight='balanced',max_depth=4)

```

```python
dt1.fit(X_train,Y_train)
y_test_pred=dt1.predict(X_test)
accuracy_score(Y_test,y_test_pred)
```

```python
dt1.fit(X_train,Y_train)
y_train_predict=dt1.predict(X_train)
accuracy_score(Y_train,y_train_predict)
```
