```python
# gradient decent -mathematics behind linear regression
```

```python
import numpy as np
import pandas as pd
```

```python
# basic function of GD
def gradient_decent(X,Y):
    m=5
    #m2=5
    #m3=5
    c=1
    learning_rate=0.01#(alpha)
    iteration=80# in case of liner regression its take by default
    n=len(X)
    for i in range(iteration):
        Y_pred=m*X+c
        cost=(1/n)*sum([value**2 for value in (Y-Y_pred)])
        dm=(-2/n)*sum((Y-Y_pred)*X)
        dc=(-2/n)*sum(Y-Y_pred)
        m=m-learning_rate*dm
        c=c-learning_rate*dc
        
        print('iteration={} ## slope={} ## intercept={} ##cost={}'.format(i+1,m,c,cost))
        
```

```python
X=np.array([1,2,3,4,5])# asume is working year
Y=np.array([20,29,31,40,50])# asume  y is salary
```

```python
gradient_decent(X,Y)
```

```python
# basic function of GD
def gradient_decent(X1,X2,X3,Y):
    m1=5
    m2=5
    m3=5
    c=1
    learning_rate=0.01
    iteration=200
    n=len(X1)
    for i in range(iteration):
        Y_pred=m1*X1+m2*X2+m3*X3+c
        cost=(1/n)*sum([value**2 for value in (Y-Y_pred)])
        dm1=(-2/n)*sum((Y-Y_pred)*X1)
        dm2=(-2/n)*sum((Y-Y_pred)*X2)
        dm3=(-2/n)*sum((Y-Y_pred)*X3)
        
        dc=(-2/n)*sum(Y-Y_pred)
        
        m1=m1-learning_rate*dm1
        m2=m2-learning_rate*dm2
        m3=m3-learning_rate*dm3  
        c=c-learning_rate*dc
        
        print('iteration={}## slope1={}## slope2={} ## slope3={} ## intercept={}##cost={}'.format(i+1,m1,m2,m3,c,cost))
        
```

```python
X1=np.array([1,2,3,4,5])
X2=np.array([1,4,3,4,5])
X3=np.array([1,1,2,6,5])
Y=np.array([20,29,31,40,50])
```

```python
gradient_decent(X1,X2,X3,Y)
```
